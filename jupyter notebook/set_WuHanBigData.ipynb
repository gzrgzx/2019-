```python
import pandas as pd 
import time

file_name = '武汉_2021_04_08迁出数据.csv'
move_out = pd.read_csv(file_name)

print(move_out.keys())
```

```python
data = [(i,j) for i,j in zip(move_out['city_name'],move_out['value'])]
data_pair = [('武汉',j) for j in move_out['city_name']]
from pyecharts import options as opts
from pyecharts.charts import Geo
from pyecharts.globals import ChartType, SymbolType

c = (
    Geo()
    .add_schema(
        maptype="china",
        itemstyle_opts=opts.ItemStyleOpts(color="#323c48", border_color="#111"),
    )
    .add(
        "",
        data,
        type_=ChartType.EFFECT_SCATTER,
        color = 'white',
    )
    .add(
        "geo",
        data_pair,
        type_=ChartType.LINES,
        effect_opts=opts.EffectOpts(
            symbol=SymbolType.ARROW, symbol_size=6
        ),
        linestyle_opts=opts.LineStyleOpts(curve=0.2),
    )
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    .set_global_opts(title_opts=opts.TitleOpts(title="武汉迁徙地图"))
)
c.render_notebook()
```
