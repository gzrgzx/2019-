```python
import pandas as pd 
import time

file_name = 'today_shanghai'+'_'+time.strftime('%Y_%m_%d',time.localtime(time.time()))+'.csv'

china_data = pd.read_csv(file_name)

print(china_data.keys())
```

```python
import pyecharts
# 调整配置项
import pyecharts.options as opts
# Map类用于绘制地图
from pyecharts.charts import Map,Page

china_data['existing_confirm'] = china_data['total_confirm'] - china_data['total_heal'] - china_data['total_dead']

china_data.head()

```

```python
for i in range(len(china_data['name'])):
    if(china_data['name'][i].find('市') == -1):
        china_data['name'][i] += '市'
       
map_data = china_data[['name','existing_confirm']].values.tolist()
map_data
```

```python
map1 = Map().add(series_name = "现存确诊人数", # 设置提示框标签
                 data_pair = map_data, # 输入数据
                 maptype = "广东", # 设置地图类型为世界地图
                 is_map_symbol_show = False # 不显示标记点
                ) 

map1.set_series_opts(label_opts=opts.LabelOpts(is_show=False))  # 不显示国家（标签）名称 
map1.set_global_opts(title_opts = opts.TitleOpts(title=" 广东省现存确诊人数地图"), # 设置图标题
                     # 设置视觉映射配置项
                     visualmap_opts = opts.VisualMapOpts())  # 显示分段式图例
map1.render_notebook()
```

```python
total_confirm = china_data[['name','total_confirm']].values.tolist()
total_confirm
```

```python
map2 = Map().add(series_name = "累计确诊人数", # 设置提示框标签
                 data_pair = total_confirm, # 输入数据
                 maptype = "广东", # 设置地图类型为世界地图
                 is_map_symbol_show = False # 不显示标记点
                ) 

map2.set_series_opts(label_opts=opts.LabelOpts(is_show=False))  # 不显示国家（标签）名称 
map2.set_global_opts(title_opts = opts.TitleOpts(title=" 广东省累计确诊人数地图"), # 设置图标题
                     # 设置视觉映射配置项
                     visualmap_opts = opts.VisualMapOpts())  # 显示分段式图例

map2.render_notebook()
```

```python
page = Page(layout=Page.SimplePageLayout)
page.add(
    map1,
    map2,
)
page.render("广东省疫情地图.html")
```
